# ORDERS CLEANING

```python
import pandas as pd
import numpy as np
```

```python
customers = pd.read_csv("olist_customers_dataset.csv")
orders = pd.read_csv("olist_orders_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
payments = pd.read_csv("olist_order_payments_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
sellers = pd.read_csv("olist_sellers_dataset.csv")
geolocation = pd.read_csv("olist_geolocation_dataset.csv")
category_translation = pd.read_csv("product_category_name_translation.csv")
```

```python
orders.head()
```

```python
orders.info()
```

```python
orders.isnull().sum()
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
```

```python
orders.info()
```

```python
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
```

```python
orders.info()
```

```python
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
```

```python
orders.info()
```

```python
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
```

```python
orders.info()
```

```python
orders['delivery_time'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days
```

```python
orders[['order_purchase_timestamp','order_delivered_customer_date','delivery_time']].head()
```

```python
orders['is_delayed'] = orders['order_delivered_customer_date'] > orders['order_estimated_delivery_date']
orders['delay_status'] = orders['is_delayed'].map({True: 'Delayed', False: 'On Time'})
```

```python
orders[['order_purchase_timestamp',
        'order_estimated_delivery_date',
        'order_delivered_customer_date',
        'delivery_time',
        'delay_status']].head()
```

```python
orders.isnull().sum()
```

```python
orders.duplicated().sum()
```

```python
orders['order_status'].value_counts()
```

```python
delivered_orders = orders[orders['order_status'] == 'delivered']
```

```python
delivered_orders.shape
```

```python
delivered_orders = delivered_orders.dropna(subset=['delivery_time'])
```

```python
delivered_orders.isnull().sum()
```

```python
delivered_orders.to_csv("cleaned_orders.csv", index=False)
```

# CUSTOMERS CLEANING

```python
customers.info()
```

```python
customers.duplicated().sum()
```

```python
customers['customers_city'] = customers['customer_city'].str.lower()
```

```python
customers['customer_city'].head()
```

```python
merged_data = orders.merge(customers, on='customer_id', how= 'left')
```

```python
merged_data.shape
```

```python
merged_data = merged_data.merge(order_items, on='order_id', how= 'left')
```

```python
merged_data.shape
```

```python
merged_data = merged_data.merge(products, on='product_id', how='left')
```

```python
merged_data.shape
```

```python
merged_data = merged_data.merge(category_translation, on='product_category_name', how='left')
```

```python
merged_data[['product_category_name', 'product_category_name_english']].head()
```

```python
merged_data = merged_data.merge(payments, on='order_id', how='left')
```

```python
merged_data.shape
```

```python
merged_data.isnull().sum()
```

```python
merged_data = merged_data.dropna(subset=['payment_value'])
```

```python
merged_data['payment_value'].isnull().sum()
```

```python
merged_data.to_csv("final_olist_cleaned_data.csv", index=False)
```
